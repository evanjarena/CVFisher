```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
s=np.load("tsense.np.npy")
p=np.load("North_pole_test/tsense.np.npy")
b=np.load("../../bao21cm/sense.npy")
bx=np.load("../../bao21cm/sense.grid.npy")/0.6761
r=np.load("../forecast_shaw/sense.npy")
```

```python
for plog in [False,True]:
    plt.figure(figsize=(12,8))
    x=np.arange(50)/10.+0.05
    plt.plot(x,1/s[:,10],'ro-',label="Adrian")
    plt.plot(x,1/p[:,10]*0.182,'go-',label="Adrian NP x 0.182")
    plt.plot(bx,1.15/b[:,100],'b-',lw=2,label="Phil x 1.15")
    plt.plot(x,1.05/r[10,:],'mo-', label="Richard x 1.05")
    plt.xlabel('$k_\perp$  $[h/{\\rm Mpc}]$',fontsize=18)
    plt.ylabel("1/tnoise", fontsize=18)
    
    plt.xlim(0.03,5)
    if plog:
        plt.semilogx()
        plt.legend(loc='upper left')
    else:
        plt.legend()
```
